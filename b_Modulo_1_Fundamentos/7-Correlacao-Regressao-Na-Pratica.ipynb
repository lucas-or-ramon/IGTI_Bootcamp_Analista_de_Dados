### Importação das Bibliotecas que serão Ultilizadas

```python
import numpy as np
from sklearn import datasets, linear_model
import matplotlib.pyplot as plt
```

### Definindos os vetores de dados e plotando a distribuição 

```python
production = np.array([877.952, 1125.295, 1317.822, 2420.78, 2625.742, 2686.193, 2916.534, 3219.187, 3368.42, 3375.228, 3518.441, 3552.19, 3617.581, 3618.559, 3800.64, 4333.917, 4809.679, 4814.012, 4970.057, 5274.943])
consumption = np.array([604.685, 682.931, 957.309, 1625.642, 2446.6, 1745.406, 2445.433, 2898.25, 1875.934, 2157.475, 2394.205, 2060.597, 2096.973, 2419.206, 2427.194, 3226.001, 2925.945, 2723.541, 3105.396, 3996.346])

plt.scatter(production, consumption, color='black')

plt.show()
```

### Definindo o modelo, fazendo o treinamento com os vetores de dados e calculando as métricas

```python
reg = linear_model.LinearRegression()
reg.fit(production.reshape(-1, 1), consumption)
y = reg.predict(production.reshape(-1, 1))

print(f'Coeficiente: {reg.coef_}')
print(f'Intercept: {reg.intercept_}')
```

### Plotando a linha de Regressão

```python
plt.scatter(production, consumption, color='black')
plt.plot(production, y, color='blue', linewidth=3)

plt.show()
```
